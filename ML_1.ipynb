# **The Job is to predict the object beneth the submarine is wheather a rock or a Mine(bomb). The submarine sends sonar singal to find the object type.**ml1.png

**What we are going to do is to collect the sonar data and feed it to the Ml model to predict**



1.   we are going to use logistic regression model as it will do well with binary data
2.   this is a supervised learning
3.   We are going to train the data in such a way that we are going to train with 90 data then the rest will done by our model





```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/Copy of sonar data.csv",header = None) # since we dont have any header we are mentioning it here
```

```python
data.head() # we have only 60 columns but python starts with the index
```

```python
data.shape # the last column says whether a rock or a mone
```

```python
data.describe()
```

```python
data[60].value_counts() # to know the number of rock and mine
```

in this data the M and R are some what equal but if the M or R is more than one another like M = 1000 and R = 500 then the predition will not be accurate

```python
# grouping the data by R and M
data.groupby(60).mean()
```

Seperating Data and the model

```python
# the labels are M and R for that we are taking two variables one for the labels and another one the data except the label
x=data.drop(columns=60,axis=1) # if we are deleting the column then the axis should be  1 and axis =0 for row droping
y=data[60]
#z=data.drop(60,axis=0) for row drop
```

```python
print(x)
print(y)
```

```python
x.shape
```

```python
y.shape
```

To split the data into training and testing the data

```python
#this order should be followed
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model training --> Logistic regression

```python
model = LogisticRegression()
```

```python
print(x_train)
print(y_train)
```

```python
# training the logistic regression model with trainging data
model.fit(x_train,y_train)
```

To check the accuracy of the model
Model Evaluation

```python
#accuracy on the traing data - if the accuracy is more than 70 % then it is good; the more the data the better accuracy
x_train_prediction=model.predict(x_train)
trainingdata_accuracy = accuracy_score(x_train_prediction, y_train) # since we are preciting for y
```

```python
print("training data accuracy :",trainingdata_accuracy ) # this means out of 100 times it can predict 83 times
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(x_test_prediction)
```

```python
print("test data accuracy :",test_data_accuracy ) # this means out of 100 times it can predict 75 times
```
